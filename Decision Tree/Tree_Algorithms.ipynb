```python
# Creating DecisionTreeClassifier with skitlearn and calculate the score
X_train, X_test, y_train, y_test = train_test_split(x, y, train_size=.25)
rf = DecisionTreeClassifier().fit(X_train, y_train)
pred_rf = rf.predict(X_test)

print(f'Score of the Random Forest: {accuracy_score(y_test, pred_rf)}')
```



*   Max Depth:

    The max depth hyperparameter in a decision tree determines the maximum depth or level of the tree. It controls the number of splits or decision nodes in the tree.



```python
# Calculate the max-depth for training size 0.25 and ploting it

X_train, X_test, y_train, y_test = train_test_split(x, y, train_size=0.25 )
scores = []

for i in range(1, 10):

    cls = DecisionTreeClassifier(max_depth=i)
    cls = cls.fit(X_train, y_train)
    X_train, _, y_train, _ = train_test_split(x, y, train_size=0.25)

    scores.append(cls.score(X_test, y_test))

num = []
for i in range(1, 10):
    num.append(i)

plt.scatter(num, scores)
plt.plot(num, scores)
plt.xlabel('Score')
plt.ylabel('Number of Depths')
plt.title('Max Depth Hyperparameter for 0.25')
plt.show()
```

In iterations (between 1 to 10), with 0.25 train size, we have changed the max-depth hyperameters and plotting it and underestand which number is best for max-depht

```python
# Calculate the max-depth for all trainig sizes of the tree and ploting it
number = 20

for j in range(1, 4):

  X_train, X_test, y_train, y_test = train_test_split(x, y, train_size=float(0.25 * j))
  scores = []

  for i in range(1, number):
     cls = DecisionTreeClassifier(max_depth=i)
     cls = cls.fit(X_train, y_train)
     X_train, _, y_train, _ = train_test_split(x, y, train_size=float(0.25 * j))

     scores.append(cls.score(X_test, y_test))

  num = []
  for i in range(1, number):
     num.append(i)

  plt.scatter(num, scores)
  plt.plot(num, scores, label = f'learning rate: {float(0.25 * j)}')

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
X_train = x
y_train = y
scores = []

for i in range(1, number):
     cls = DecisionTreeClassifier(max_depth=i)
     cls = cls.fit(X_train, y_train)
     scores.append(cls.score(X_test, y_test))

num = []
for i in range(1, number):
     num.append(i)

plt.xlabel('Number of iterates')
plt.ylabel('Test dcore')
plt.title(f'Max Depth Hyperparameter')
plt.scatter(num, scores)
plt.plot(num, scores, label = f'learning rate: 1')
plt.legend()
plt.show()
```

In iterations (between 1 to 20), with Different training sizes, we have changed the max-depth hyperameters and plotting it and underestand which number is best for max-depht



*   Min Samples Leaf:

    The min samples leaf hyperparameter in a decision tree specifies the minimum number of samples required to be at a leaf node. A leaf node is a node that does not split further and represents a final decision or prediction.


```python
# Calculate the min samples leaf for training size 0.25 and ploting it

X_train, X_test, y_train, y_test = train_test_split(x, y, train_size=0.25 )
scores = []

for i in range(1, 10):

    cls = DecisionTreeClassifier(min_samples_leaf=i)
    cls = cls.fit(X_train, y_train)
    X_train, _, y_train, _ = train_test_split(x, y, train_size=0.25)

    scores.append(cls.score(X_test, y_test))

num = []
for i in range(1, 10):
    num.append(i)

plt.scatter(num, scores)
plt.plot(num, scores)
plt.xlabel('Score')
plt.ylabel('Number of Depths')
plt.title('Min Samples leaf Hyperparameter for 0.25')
plt.show()
```

In iterations (between 1 to 10), with 0.25 training sizes, we have changed the Min Samples leaf hyperameters and plotting it and underestand which number is best for Min Samples leaf

```python
# Calculate the min samples leaf for all trainig sizes of the tree and ploting it

number = 30

for j in range(1, 4):

  X_train, X_test, y_train, y_test = train_test_split(x, y, train_size=float(0.25 * j))
  scores = []

  for i in range(1, number):
     cls = DecisionTreeClassifier(min_samples_leaf=i)
     cls = cls.fit(X_train, y_train)
     X_train, _, y_train, _ = train_test_split(x, y, train_size=float(0.25 * j))

     scores.append(cls.score(X_test, y_test))

  num = []
  for i in range(1, number):
     num.append(i)

  plt.scatter(num, scores)
  plt.plot(num, scores, label = f'learning rate: {float(0.25 * j)}')

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
X_train = x
y_train = y
scores = []

for i in range(1, number):
     cls = DecisionTreeClassifier(max_depth=i)
     cls = cls.fit(X_train, y_train)
     scores.append(cls.score(X_test, y_test))

num = []
for i in range(1, number):
     num.append(i)

plt.xlabel('Number of iterates')
plt.ylabel('Test dcore')
plt.title(f'Min Samples Leaf Hyperparameter')
plt.scatter(num, scores)
plt.plot(num, scores, label = f'learning rate: 1')
plt.legend()
plt.show()
```

In iterations (between 1 to 20), with Different training sizes, we have changed the Min Samples Leaf hyperameters and plotting it and underestand which number is best for Min Samples Leaf

```python
# This function use for drawing the learning curve
def draw_plot(Classifier, name):
  train_sizes, train_scores, validation_scores = learning_curve(
  estimator = Classifier,
  X = x,
  y = y,
  train_sizes = [.25, .5, .75, 1],
  cv = 5,
  scoring = 'accuracy')

  train_scores_mean = np.mean(train_scores, axis=1)
  train_scores_std = np.std(train_scores, axis=1)
  test_scores_mean = np.mean(validation_scores, axis=1)
  test_scores_std = np.std(validation_scores, axis=1)

  # Plot the learning curve
  plt.figure()
  plt.title(f"Learning Curve for {name}")
  plt.xlabel("Training Examples")
  plt.ylabel("Score")

  # Plot training set scores
  plt.plot(train_sizes, train_scores_mean, 'o-', color="b", label="Training score")
  plt.fill_between(train_sizes, train_scores_mean - train_scores_std,
                  train_scores_mean + train_scores_std, alpha=0.1, color="r")

  # Plot cross-validation scores
  plt.plot(train_sizes, test_scores_mean, 'o-', color="r", label="Testing score")
  plt.fill_between(train_sizes, test_scores_mean - test_scores_std,
                  test_scores_mean + test_scores_std, alpha=0.1, color="g")

  # Add legend and display plot
  plt.legend(loc="best")
  plt.grid()
  plt.xlim(0, 4500)
  plt.show()
```

Drawing RANDOM FOREST learning curve.


---

The training score is always one for each training size. It's kind of Tree.
Due to the learning rates with training of 0.25, 0.5, 0.75, and 1, the testing score is between 0.68 and 0.72.

```python
draw_plot(RandomForestClassifier(), "Random Forest")
```

The training score decreases when the training size is increased.
Due to the learning rates with training of 0.25, 0.5, 0.75, and 1, the testing score is between 0.67 and 0.72.
